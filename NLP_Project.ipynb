```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline


with open('clean.txt', 'r', encoding='utf-8') as f:
    swahili_text = f.read().splitlines()

# Splitin text to words
swahili_words = [word for line in swahili_text for word in line.split()]

text_classification = Pipeline([
    ('vect', CountVectorizer()), 
    ('tfidf', TfidfTransformer()),
    ('clf', MultinomialNB())
])

text_classification.fit(swahili_words, ['positive']*len(swahili_words))

# evaluating sentiment of the words
sentiment = text_classification.predict(swahili_words)

# Printing sentiment of the words
for word, sent in zip(swahili_words, sentiment):
    print(word,sent)
```

```python
import nltk

stop_words = [
    'na', 'ya', 'kwa', 'za', 'ni', 'wakati', 'hata', 'ili', 'kuwa', 
    'au', 'katika', 'kama', 'ambayo', 'yake', 'kwa sababu', 'hapo awali', 
    'kila', 'bila', 'baada ya', 'ingawa', 'pia', 'ndani', 'juu ya', 
    'kwa hiyo', 'mwisho', 'hivyo', 'kutoka', 'hapo', 'mara', 'kabla',
    'kutokana', 'kuelekea', 'kati ya', 'zaidi', 'chini ya', 'leo', 
    'kwenda', 'mpaka', 'pengine', 'hapana', 'huyo', 'pamoja',  
    'kiasi', 'labda', 'chache', 'yangu', 'yako', 'yao',
   ]

# Load the dataset
dataset = [("sentence1", "positive"), ("sentence2", "negative"), ("sentence3", "neutral"), ...]

# Tokenize the sentences and remove stop words
sentences = []
for (sentence, sentiment) in dataset:
    words = nltk.word_tokenize(sentence.lower())
    words = [word for word in words if word not in stop_words]
    sentences.append((words, sentiment))

def EX_features(words):
    return dict([(word, True) for word in words])

featuresets = [(EX_features(words), sentiment) for (words, sentiment) in sentences]

# Spliting training and testing sets
train_set, test_set = featuresets[:int(len(featuresets)*0.8)], featuresets[int(len(featuresets)*0.8):]

# Train a Naive Bayes classifier on the training set
classifier = nltk.NaiveBayesClassifier.train(train_set)

# new sentences
for sentence in test_sentences:
    features = EX_features(nltk.word_tokenize(sentence.lower()))
    sentiment = classifier.classify(features)
    print(sentence, sentiment)
```
